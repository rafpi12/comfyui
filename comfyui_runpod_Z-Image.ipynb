```python
# 🚀 DÉMARRAGE V21.18 - ULTIMATE COMMANDER (NVVFX FULL FIX)
import os, subprocess, sys, shutil, torch, threading, time, re, json

# --- CONFIGURATION ---
COMFY_DIR = "/workspace/ComfyUI"
VENV_DIR = "/workspace/my_env"
PYTHON_EXEC = f"{VENV_DIR}/bin/python"
PIP_EXEC = f"{VENV_DIR}/bin/pip"
SENTINEL = f"{VENV_DIR}/.install_done_v2118"
NVVFX_LIBS = f"{VENV_DIR}/lib/python3.11/site-packages/nvvfx/libs"
EXTRA_LIBS = f"{VENV_DIR}/extra_libs"
NGX_DENY_LIST = "/usr/share/nvidia/ngx/config/versions/1/files/Linux_amd64_nvngx_deny_list.txt"
LDCONF_NVVFX = "/etc/ld.so.conf.d/nvvfx.conf"
NGX_DL_URL = "https://us.download.nvidia.com/XFree86/Linux-x86_64/565.57.01/NVIDIA-Linux-x86_64-565.57.01.run"

os.environ["PIP_ROOT_USER_ACTION"] = "ignore"
os.environ["PYTHONNOUSERSITE"] = "1"
os.environ["DEBIAN_FRONTEND"] = "noninteractive"
os.environ["MPLBACKEND"] = "Agg"

print("💎 DÉMARRAGE V21.18 - ULTIMATE COMMANDER (NVVFX FULL FIX & 4090 OPTIMIZED)")

# ==================================================================================
# CACHE GLOBAL & UTILS
# ==================================================================================
_INSTALLED_PKGS_CACHE = None

def get_installed_packages_bulk():
    global _INSTALLED_PKGS_CACHE
    if _INSTALLED_PKGS_CACHE is not None: return _INSTALLED_PKGS_CACHE
    if not os.path.exists(PYTHON_EXEC): return set()
    result = subprocess.run([PYTHON_EXEC, "-m", "pip", "list", "--format=json"], capture_output=True, text=True)
    try:
        packages = json.loads(result.stdout)
        _INSTALLED_PKGS_CACHE = {pkg["name"].lower().replace("-", "_") for pkg in packages}
    except: _INSTALLED_PKGS_CACHE = set()
    return _INSTALLED_PKGS_CACHE

def invalidate_pkg_cache():
    global _INSTALLED_PKGS_CACHE
    _INSTALLED_PKGS_CACHE = None

def run_pip_visible(args_list, no_isolation=False):
    base_cmd = [PYTHON_EXEC, "-m", "pip", "install", "--no-cache-dir", "--prefer-binary"]
    if no_isolation: base_cmd.append("--no-build-isolation")
    subprocess.check_call(base_cmd + args_list)
    invalidate_pkg_cache()

def pip_install_if_missing(packages):
    installed_cache = get_installed_packages_bulk()
    to_install = [pkg for pkg in packages if re.split(r'[>=<!;\[]', pkg)[0].strip().lower().replace("-", "_") not in installed_cache]
    if to_install:
        print(f"    -> 📦 Install/check ({len(to_install)}) : {', '.join(to_install[:5])}")
        run_pip_visible(to_install)
    else: print("    -> ✅ Librairies Python OK.")

# ==================================================================================
# INSTALLATION & PATCHING (SENTINEL CHECK)
# ==================================================================================
if os.path.exists(SENTINEL):
    print("\n⚡ Installation déjà effectuée (v21.18) — skip sections 1 à 4.")
else:
    print("\n🔧 [1/5] Préparation système...")
    sys_pkgs = ["ncdu", "ffmpeg", "git", "cmake", "aria2", "libsndfile1-dev", "build-essential"]
    subprocess.run(["apt-get", "update"], stdout=subprocess.DEVNULL)
    subprocess.run(["apt-get", "install", "-y", "--no-upgrade"] + sys_pkgs, stdout=subprocess.DEVNULL)

    print("\n📦 [2/5] Installation de ComfyUI-NAG...")
    nag_path = f"{COMFY_DIR}/custom_nodes/ComfyUI-NAG"
    if not os.path.exists(nag_path):
        subprocess.run(["git", "clone", "https://github.com/ChenDarYen/ComfyUI-NAG.git", nag_path])

    if not os.path.exists(VENV_DIR):
        subprocess.check_call([sys.executable, "-m", "venv", VENV_DIR])

    print("\n🐍 [3/5] Librairies Python...")
    run_pip_visible(["--upgrade", "pip", "wheel", "setuptools"])
    pypi_packages = ["einops", "wand", "transformers>=4.48.0", "accelerate", "scipy", "insightface==0.7.3", "onnxruntime-gpu", "opencv-contrib-python", "google-generativeai", "google-genai", "anthropic", "vtracer", "xformers"]
    pip_install_if_missing(pypi_packages)

    print("\n🩹 [4/5] Patches de compatibilité...")
    mm_path = f"{COMFY_DIR}/comfy/model_management.py"
    if os.path.exists(mm_path):
        with open(mm_path, 'r') as f: content = f.read()
        if "cu130" in content:
            with open(mm_path, 'w') as f: f.write(content.replace("cu130", "cu128"))
            print("    -> ✅ Patch cu128 appliqué.")

    open(SENTINEL, 'w').close()

# ==================================================================================
# MAINTENANCE & HOTFIXES (VÉRIFIÉS À CHAQUE BOOT)
# ==================================================================================
print("\n🛠️ [HOTFIX] Application des correctifs NAG & RTX...")

# --- 1. Fix Imports NAG model.py ---
nag_model_py = f"{COMFY_DIR}/custom_nodes/ComfyUI-NAG/chroma/model.py"
if os.path.exists(nag_model_py):
    subprocess.run(["git", "checkout", "chroma/model.py"],
        cwd=f"{COMFY_DIR}/custom_nodes/ComfyUI-NAG",
        stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    with open(nag_model_py, 'r') as f: content = f.read()
    old = """from comfy.ldm.chroma.layers import (
    DoubleStreamBlock,
    SingleStreamBlock,
)
from comfy.ldm.chroma.model import Chroma"""
    new = """try:
    from comfy.ldm.chroma.layers import DoubleStreamBlock, SingleStreamBlock
    from comfy.ldm.chroma.model import Chroma
except (ImportError, AttributeError):
    from comfy.ldm.flux.layers import DoubleStreamBlock, SingleStreamBlock
    from comfy.ldm.flux.model import Flux as Chroma"""
    if old in content:
        with open(nag_model_py, 'w') as f: f.write(content.replace(old, new))
        print("    -> ✅ NAG model.py : Import chroma→flux patché.")
    else:
        print("    -> ⚠️  NAG model.py : Bloc cible introuvable (déjà patché?).")

# --- 2. Patch Imports NAG layers.py ---
nag_layers_py = f"{COMFY_DIR}/custom_nodes/ComfyUI-NAG/chroma/layers.py"
if os.path.exists(nag_layers_py):
    with open(nag_layers_py, 'r') as f: content = f.read()
    correct_import = "from comfy.ldm.flux.layers import DoubleStreamBlock, SingleStreamBlock"
    new_content = re.sub(r'from comfy\.ldm\.chroma\.layers import.*?(?=import|class|def|$)',
        correct_import + "\n", content, flags=re.DOTALL)
    if new_content != content:
        with open(nag_layers_py, 'w') as f: f.write(new_content)
        print("    -> ✅ NAG layers.py : Imports Flux synchronisés.")

# --- 3. NvVFX : ldconfig + libnvidia-ngx + NGX config ---
print("\n🔧 [HOTFIX] NvVFX setup...")

# ldconfig pour nvvfx/libs et extra_libs
os.makedirs(EXTRA_LIBS, exist_ok=True)
ldconf_content = f"{NVVFX_LIBS}\n{EXTRA_LIBS}\n"
current_ldconf = open(LDCONF_NVVFX).read() if os.path.exists(LDCONF_NVVFX) else ""
if current_ldconf != ldconf_content:
    with open(LDCONF_NVVFX, 'w') as f: f.write(ldconf_content)
    subprocess.run(["ldconfig"], stdout=subprocess.DEVNULL)
    print("    -> ✅ ldconfig nvvfx mis à jour.")
else:
    print("    -> ✅ ldconfig nvvfx OK.")

# libnvidia-ngx.so depuis le runfile driver (si absent)
ngx_so = f"{EXTRA_LIBS}/libnvidia-ngx.so.565.57.01"
ngx_so1 = f"{EXTRA_LIBS}/libnvidia-ngx.so.1"
if not os.path.exists(ngx_so):
    print("    -> 📥 Téléchargement libnvidia-ngx depuis le runfile driver (~320MB)...")
    run_path = "/tmp/NVIDIA-Linux-x86_64-565.57.01.run"
    extract_path = "/tmp/nvidia_ngx_extract"
    subprocess.run(["wget", "-q", "--show-progress", "-O", run_path, NGX_DL_URL], check=True)
    subprocess.run(["chmod", "+x", run_path])
    subprocess.run([run_path, "--extract-only", f"--target={extract_path}"],
        stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    src = f"{extract_path}/libnvidia-ngx.so.565.57.01"
    if os.path.exists(src):
        shutil.copy2(src, ngx_so)
        if not os.path.exists(ngx_so1):
            os.symlink(ngx_so, ngx_so1)
        subprocess.run(["ldconfig"], stdout=subprocess.DEVNULL)
        print("    -> ✅ libnvidia-ngx.so installé.")
    else:
        print("    -> ⚠️  libnvidia-ngx.so introuvable dans le runfile.")
    shutil.rmtree(extract_path, ignore_errors=True)
    os.remove(run_path) if os.path.exists(run_path) else None
else:
    if not os.path.exists(ngx_so1):
        os.symlink(ngx_so, ngx_so1)
        subprocess.run(["ldconfig"], stdout=subprocess.DEVNULL)
    print(f"    -> ✅ libnvidia-ngx.so OK.")

# NGX config deny_list (nécessaire pour l'init NGX)
os.makedirs(os.path.dirname(NGX_DENY_LIST), exist_ok=True)
if not os.path.exists(NGX_DENY_LIST):
    open(NGX_DENY_LIST, 'w').close()
    print("    -> ✅ NGX deny_list créé.")
else:
    print("    -> ✅ NGX deny_list OK.")

# Test NvVFX rapide (non-bloquant)
nvvfx_ok = False
try:
    result = subprocess.run([PYTHON_EXEC, "-c",
        "import nvvfx; sr = nvvfx.VideoSuperRes(); sr.output_width=1920; sr.output_height=1080; sr.load(); print('OK')"],
        capture_output=True, text=True, timeout=15)
    if "OK" in result.stdout:
        nvvfx_ok = True
        print("    -> ✅ NvVFX VideoSuperRes : opérationnel !")
    else:
        print(f"    -> ⚠️  NvVFX non disponible sur cet hôte (pod incompatible NGX).")
except Exception as e:
    print(f"    -> ⚠️  NvVFX test skipped : {e}")

# --- 4. Paths SDK ---
IM7_PREFIX = "/usr/local"
nvidia_libs = "/usr/local/cuda/lib64:/usr/lib/x86_64-linux-gnu/nvidia:/usr/lib/x86_64-linux-gnu"
os.environ["MAGICK_HOME"] = IM7_PREFIX
os.environ["LD_LIBRARY_PATH"] = f"{NVVFX_LIBS}:{EXTRA_LIBS}:{nvidia_libs}:{IM7_PREFIX}/lib:" + os.environ.get("LD_LIBRARY_PATH", "")

# ==================================================================================
# 5. LANCEMENT (4090 OPTIMIZED)
# ==================================================================================
print("\n🚀 [5/5] Lancement ComfyUI (Mode HighVRAM)...")
runpod_id = os.environ.get("RUNPOD_POD_ID")
if runpod_id: print(f"\n🔗 LIEN RUNPOD : https://{runpod_id}-8188.proxy.runpod.net")

run_script = f"""#!/bin/bash
source {VENV_DIR}/bin/activate
export MAGICK_HOME="{IM7_PREFIX}"
export LD_LIBRARY_PATH="{NVVFX_LIBS}:{EXTRA_LIBS}:{nvidia_libs}:{IM7_PREFIX}/lib:$LD_LIBRARY_PATH"
export PYTHONNOUSERSITE=1
export MPLBACKEND=Agg
cd {COMFY_DIR}
python main.py --listen 0.0.0.0 --port 8188 --highvram --enable-cors-header --preview-method auto
"""
with open("run_comfy.sh", "w") as f: f.write(run_script)
os.chmod("run_comfy.sh", 0o755)
get_ipython().system('./run_comfy.sh')
```
