```python
import os
import shutil
import configparser
import re
import subprocess
import time

# --- CONFIGURATION ---
COMFY_DIR = "/workspace/ComfyUI"
CUSTOM_NODES = f"{COMFY_DIR}/custom_nodes"
PULID_DIR = f"{CUSTOM_NODES}/ComfyUI_PuLID_Flux_ll"
MANAGER_DIR = f"{CUSTOM_NODES}/ComfyUI-Manager"

print("ðŸš€ DÃ‰MARRAGE DU SCRIPT TOUT-EN-UN (Installation & RÃ©paration)...")

# --- 1. DÃ‰VERROUILLAGE DU MANAGER ---
print("\nðŸ”“ [1/6] DÃ©verrouillage du Manager...")
config_paths = [
    f"{COMFY_DIR}/user/default/ComfyUI-Manager/config.ini",
    f"{MANAGER_DIR}/config.ini"
]
for path in config_paths:
    if not os.path.exists(os.path.dirname(path)):
        os.makedirs(os.path.dirname(path), exist_ok=True)
    config = configparser.ConfigParser()
    if os.path.exists(path): config.read(path)
    if 'default' not in config: config['default'] = {}
    config['default']['security_level'] = 'weak'
    with open(path, 'w') as f: config.write(f)

# --- 2. INSTALLATION SYSTÃˆME ---
print("\nðŸ”§ [2/6] Installation outils systÃ¨me (ffmpeg, aria2)...")
subprocess.run(["apt-get", "update"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
subprocess.run(["apt-get", "install", "-y", "ffmpeg", "aria2"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# --- 3. DÃ‰PENDANCES PYTHON (La partie critique) ---
print("\nðŸ“¦ [3/6] Installation des librairies Python...")
# Nettoyage prÃ©ventif
subprocess.run(["pip", "uninstall", "-y", "xformers", "torch", "torchvision", "torchaudio", "huggingface_hub", "transformers", "diffusers", "insightface"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# PyTorch Stable
print("   -> PyTorch...")
subprocess.run(["pip", "install", "torch==2.5.1+cu124", "torchvision==0.20.1+cu124", "torchaudio==2.5.1+cu124", "--index-url", "https://download.pytorch.org/whl/cu124"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# Stack InsightFace & PuLID & Bases
print("   -> DÃ©pendances ComfyUI & PuLID...")
subprocess.run(["pip", "install", "--upgrade", "opencv-python", "GitPython", "scikit-image", "onnxruntime-gpu", "chardet", "matplotlib", "toml", "piexif", "rich", "timm", "ultralytics", "wget", "watchdog", "imageio-ffmpeg", "pickleshare", "insightface", "facexlib"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
subprocess.run(["pip", "install", "transformers>=4.46.3", "huggingface-hub>=0.26.5", "diffusers>=0.31.0", "accelerate"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# FIX BUG AIOHTTP/YARL (Le crash 'ValueError: Host') + Autres modules
print("   -> Correction bug connexion (aiohttp/yarl)...")
subprocess.run(["pip", "install", "--upgrade", "aiohttp", "yarl", "av", "einops", "torchsde", "kornia", "spandrel", "soundfile", "moviepy"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# --- 4. NETTOYAGE & INSTALLATION PuLID (lldacing) ---
print("\nðŸ§¹ [4/6] Installation de PuLID (Version lldacing)...")
# Nettoyage de TOUS les dossiers PuLID existants pour Ã©viter les conflits
for folder in os.listdir(CUSTOM_NODES):
    if "pulid" in folder.lower():
        path = os.path.join(CUSTOM_NODES, folder)
        print(f"   ðŸ—‘ï¸ Suppression : {folder}")
        shutil.rmtree(path, ignore_errors=True)

# Clonage
print(f"   ðŸ“¥ Clonage de {PULID_DIR}...")
!git clone https://github.com/lldacing/ComfyUI_PuLID_Flux_ll.git {PULID_DIR}

print("   ðŸ“¦ Installation requirements spÃ©cifiques...")
subprocess.run(["pip", "install", "-r", f"{PULID_DIR}/requirements.txt"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
subprocess.run(["pip", "install", "facenet-pytorch", "--no-deps"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# PATCH DU CODE (Fix 'latent_shapes')
print("   ðŸ’‰ Patching du code PuLID pour ComfyUI v0.3.76+...")
patched = False
for fname in ["pulid.py", "nodes.py", "pulid_flux.py"]:
    fpath = os.path.join(PULID_DIR, fname)
    if os.path.exists(fpath):
        with open(fpath, 'r') as f: content = f.read()
        # Regex : cherche la fonction et ajoute **kwargs
        pattern = r"(def\s+pulid_outer_sample_wrappers_with_override\s*\([^)]*?)(\)\s*:)"
        if re.search(pattern, content):
            new_content = re.sub(pattern, r"\1, **kwargs\2", content, count=1)
            with open(fpath, 'w') as f: f.write(new_content)
            print(f"   âœ… Patch appliquÃ© sur {fname} !")
            patched = True
            break
        elif "**kwargs" in content and "pulid_outer_sample_wrappers_with_override" in content:
             print(f"   â„¹ï¸ {fname} est dÃ©jÃ  compatible.")
             patched = True

if not patched:
    print("   âš ï¸ ATTENTION : Impossible de trouver la fonction Ã  patcher.")

# --- 5. TÃ‰LÃ‰CHARGEMENT DES MODÃˆLES ---
print("\nðŸ“¥ [5/6] VÃ©rification des ModÃ¨les...")

# InsightFace AntelopeV2 (Miroir public Okaris)
antelope_dir = f"{COMFY_DIR}/models/insightface/models/antelopev2"
os.makedirs(antelope_dir, exist_ok=True)
files_antelope = ["1k3d68.onnx", "2d106det.onnx", "genderage.onnx", "glintr100.onnx", "scrfd_10g_bnkps.onnx"]
print("   -> ModÃ¨les InsightFace...")
for f in files_antelope:
    if not os.path.exists(f"{antelope_dir}/{f}"):
        subprocess.run(["aria2c", "-x", "16", "-s", "16", "-k", "1M", "-c", f"https://huggingface.co/okaris/antelopev2/resolve/main/{f}?download=true", "-d", antelope_dir, "-o", f], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# PuLID Model
pulid_flux_dir = f"{COMFY_DIR}/models/pulid"
os.makedirs(pulid_flux_dir, exist_ok=True)
if not os.path.exists(f"{pulid_flux_dir}/pulid_flux_v0.9.0.safetensors"):
    print("   -> ModÃ¨le PuLID Flux...")
    subprocess.run(["aria2c", "-x", "16", "-s", "16", "-k", "1M", "-c", "https://huggingface.co/guozinan/PuLID/resolve/main/pulid_flux_v0.9.0.safetensors", "-d", pulid_flux_dir, "-o", "pulid_flux_v0.9.0.safetensors"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# Eva Clip
eva_dir = f"{COMFY_DIR}/models/clip"
if not os.path.exists(f"{eva_dir}/EVA02_CLIP_L_336_psz14_s6B.pt"):
    print("   -> ModÃ¨le Eva Clip...")
    subprocess.run(["aria2c", "-x", "16", "-s", "16", "-k", "1M", "-c", "https://huggingface.co/QuanSun/EVA-CLIP/resolve/main/EVA02_CLIP_L_336_psz14_s6B.pt", "-d", eva_dir, "-o", "EVA02_CLIP_L_336_psz14_s6B.pt"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# --- 6. LANCEMENT ---
print("\nðŸš€ [6/6] Lancement de ComfyUI...")

# Nettoyage WanVideo
if os.path.exists(f"{CUSTOM_NODES}/ComfyUI-wanBlockswap"):
    shutil.move(f"{CUSTOM_NODES}/ComfyUI-wanBlockswap", f"{CUSTOM_NODES}/ComfyUI-wanBlockswap_DISABLED")

# Script de lancement
run_script = """#!/bin/bash
cd /workspace/ComfyUI
echo "ðŸ”¥ Lancement OptimisÃ© (BF16)..."
python main.py --listen 0.0.0.0 --port 8188 --disable-auto-launch --fast
"""
with open("run_comfy.sh", "w") as f: f.write(run_script)
os.chmod("run_comfy.sh", 0o755)

print("âœ… TERMINÃ‰. Le serveur dÃ©marre ci-dessous...")
!./run_comfy.sh
```
